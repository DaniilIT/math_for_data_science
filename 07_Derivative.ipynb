# Производная

```python
from IPython.display import display, Latex
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def my_plot(f):
    """
    Показывает на первом графике функцию f, на втором - её производную
    """
    def function(func, x_):
        return float(func.subs(x, x_))
    
    f_derivative = f.diff(x)
    
    X = np.linspace(-10, 10, num=200)
    y1 = [function(f, x_) for x_ in X]
    y2 = [function(f_derivative, x_) for x_ in X]
    
    fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=[12, 8])
    
    fig.suptitle('function and its derivative')
    
    ax1.set_title('f(x) = ' + str(f))
    ax1.plot(X, y1, 'b')
    
    ax2.set_title('f\'(x) = ' + str(f_derivative))
    ax2.plot(X, y2, 'g')
    ax2.axhline(0, color='black', linewidth=1)
    
    plt.show();
```

```python
x = sp.Symbol('x')
```

$$
(\upsilon + \nu)' = \upsilon' + \nu'
$$

$$
(c f(x))' = c f'(x)
$$

Степенная функция:
$$
(x^n)' = n x^{n-1}
$$

Показательная функция:
$$
(a^x)' = a^x \ln{a} \quad (e^x)' = e^x
$$

$$
(\log_{a}{x})' = \frac{1}{x \ln{a}} \quad (\ln{x})' = \frac{1}{x}
$$

$$
(\sin{x})' = \cos{x} \quad (\cos{x})' = -\sin{x}
$$

```python
f = x**3 - 2*x**2 + x - 6
f_derivative = f.diff(x)
tex = "$ ({})' = {} $".format(sp.latex(f), sp.latex(f_derivative))
display(Latex(tex))
```

```python
my_plot(f)
```

```python
roots = sp.solve(f_derivative)
for i, v in enumerate(roots):
    roots[i] = v.evalf(5)
roots
```

```python
f = sp.pi**x - 9/2 * sp.exp(x) + 2/3 * sp.sin(x)
f = sp.nsimplify(f) #  оставить дробь
f_derivative = f.diff(x)
tex = "$ ({})' = {} $".format(sp.latex(f), sp.latex(f_derivative))
display(Latex(tex))
```

```python
my_plot(f)
```

$$
(\upsilon \cdot \nu)' = \upsilon' \nu + \upsilon \nu'
$$

$$
\left( \frac{\upsilon}{\nu} \right)' = \frac{\upsilon' \nu - \upsilon \nu'}{\nu^2}
$$

$$
(f(g(x)))' = f'(g(x)) \cdot g'(x)
$$

```python
f = 3 * x**2 * sp.sin(x)
f_derivative = f.diff(x)
tex = "$ ({})' = {} $".format(sp.latex(f), sp.latex(f_derivative))
display(Latex(tex))
```

```python
my_plot(f)
```
