```python
import sympy  as sp
from sympy.plotting import plot, plot3d
sp.init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = sp.symbols('x, y')
f = 5*x - 2*y + 1
f
```

```python
f.subs({x:2, y:5})
```

```python
plot3d(f);
```

```python
roots = sp.solve(f)
roots
```

```python
x = roots[0][x]
x
```

```python
plot(x);
```

```python
x, y = sp.symbols('x, y')
f = x**2 + y**2 - 1
f
```

```python
plot3d(f);
```

```python
roots = sp.solve(f)
roots
```

```python
x1 = roots[0][x]
x2 = roots[1][x]
```

```python
p = plot(x1, x2, title='roots', xlabel='x', ylabel='y', size=(6, 6), show=False)
p[0].line_color = 'g'
p[1].line_color = 'b'
p.show();
```
