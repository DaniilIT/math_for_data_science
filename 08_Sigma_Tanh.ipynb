```python
import sympy as sp
from sympy.plotting import plot
```

```python
x = sp.Symbol('x')
```

## Sigma

$ \sigma(x) = \cfrac{1}{1 + e^{-x}} $ - сигмоида - логистическая функция

```python
f = 1 / (1 + sp.exp(-x))
f
```

```python
plot(f);
```

$ \sigma'(x) = \left( \cfrac{1}{1 + e^{-x}} \right)' =
\cfrac{1' (1 + e^{-x}) - 1 (1 + e^{-x})'}{(1 + e^{-x})^2} =
\cfrac{e^{-x}}{(1 + e^{-x})^2} =$ \
$ = \cfrac{e^{-x} + 1 - 1}{1 + e^{-x}} \cdot \cfrac{1}{1 + e^{-x}} =
\left( \cfrac{e^{-x} + 1}{1 + e^{-x}} - \cfrac{1}{1 + e^{-x}} \right) 
\cdot \cfrac{1}{1 + e^{-x}} =
\cfrac{1}{1 + e^{-x}} \cdot \left( 1 - \cfrac{1}{1 + e^{-x}} \right) =
\sigma(x) \cdot (1 - \sigma(x)) $

```python
f_derivative = f.diff(x)
sp.cancel(f_derivative) # придать рациональный вид дроби
f_derivative
```

```python
f_derivative = sp.cancel(f * (1 - f))
f_derivative
```

Производная сигмоиды может быть легко выражена через саму функцию, что позволяет существенно сократить вычислительную сложность метода обратного распространения ошибки.

```python
plot(f_derivative);
```

## Tanh

$ \tanh(x) = \cfrac{e^x - e^{-x}}{e^x + e^{-x}} $ - гиперболический (выражается через экспоненту) тангенс - логистическая функция

```python
f = (sp.exp(x) - sp.exp(-x)) / (sp.exp(x) + sp.exp(-x))
f
```

```python
plot(f);
```

$ \tanh'(x) = \left( \cfrac{e^x - e^{-x}}{e^x + e^{-x}} \right)' =
\cfrac{(e^x - e^{-x})' (e^x + e^{-x}) - (e^x - e^{-x}) (e^x + e^{-x})'}{(e^x + e^{-x})^2} = $
$ = \cfrac{(e^x + e^{-x}) (e^x + e^{-x}) - (e^x - e^{-x}) (e^x - e^{-x})}{(e^x + e^{-x})^2} =
\cfrac{(e^x + e^{-x})^2}{(e^x + e^{-x})^2} - \cfrac{(e^x - e^{-x})^2}{(e^x + e^{-x})^2} =
1 - \left( \cfrac{e^x - e^{-x}}{e^x + e^{-x}} \right)^2 = 1 - \tanh^2(x) $

```python
f_derivative = f.diff(x)
f_derivative = sp.collect(f_derivative, sp.exp(x) - sp.exp(-x)) # вынести за скобку
f_derivative
```

```python
f_derivative = 1 - f**2
f_derivative
```

```python
plot(f_derivative);
```
